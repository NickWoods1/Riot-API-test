```python
# Get champion IDs
with open('champion_ids.txt') as champ_ids:
    champs = {}
    for i, line in enumerate(champ_ids):
        for j, char in enumerate(line):
            if char == ':':
                champs[f'{line[j+1:]}'.strip()] = int(line[:j])
```

```python
with open('plat_champion_winrates.txt') as plat_winrates:
    winrates = []
    for i, line in enumerate(plat_winrates):
        if (i - 2) % 3 == 0:
            for j, char in enumerate(line):
                if char == '%':
                    winrates.append(line[:j])

champ_winrates = {}
for champion in winrates:
    champ, winrate = champion.split('\t')[0], float(champion.split('\t')[1])
    """
    if champ == 'Nunu & Willump':
        champ = 'Nunu'    
    if champ == "Kha'Zix":
        champ = 'Khazix'
    if champ == "Kai'Sa":
        champ = 'Kaisa'
    if champ == "Rek'Sai":
        champ = 'Reksai'
    if champ == "Kog'Maw":
        champ = 'Kogmaw'
    if champ == "Cho'Gath":
        champ = 'Chogath'
    if champ == "Vel'Koz":
        champ = 'Velkoz'
    if champ == "Master Yi":
        champ = 'MasterYi'
    """
    champ_winrates[champ] = winrate
```

```python
# Turn dict champ winrates into dataframe and save to csv for future use
a, b = [], []
for champ in champ_winrates:
    a.append(champ)
    b.append(champ_winrates[champ])

x = pd.DataFrame(data={'Champion': a, 'Winrate': b}, columns=['Champion','Winrate'])
x.to_csv('plat_champ_winrates.csv')
```
